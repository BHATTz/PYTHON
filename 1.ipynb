```python
n1 = int(input("Enter First number :"))
n2 = int(input("Enter Second number :"))
x = n1
y = n2
while(n2!=0):
    t = n2
    n2 = n1 % n2
    n1 = t
gcd = n1
print("GCD of {0} and {1} = {2}".format(x,y,gcd))
lcm = (x*y)/gcd
print("LCM of {0} and {1} = {2}".format(x,y,lcm))
```

```python
n=input("Enter a number : ")
sum=0
for i in range(0,len(n)):
    sum=sum+pow(int(n[i]),3)

print("Sum of cube of digits  : ",sum)
if(sum==int(n)):
    print("Digits are matching to cube  : ", n)
else:
    print("Digits are NOT matching to cube : ", n)

my_list=[]
for n in range(100,1000):
    sum=0
    my_str=str(n)
    k=len(my_str)
    for i in range(0,k):
        sum=sum+pow(int(my_str[i]),3)
    if(sum==int(n)):
        my_list.append(n)
print("Highest number : ",max(my_list))
print("Lowest number : ",min(my_list))
```

```python
def bubblesorting(nlist):
    for num in range(len(n)-1,0,-1):
        for i in range(num):
            if n[i]>nlist[i+1]:
                temp = n[i]
                n[i] = n[i+1]
                n[i+1] = temp

n=[69,88,48,99,19,35,45]
bubblesorting(n)
print(n)
```

```python
#Simple stacked bar plot
import matplotlib.pyplot as plt
import numpy as np
x = ['A','B','C','D']
y1 = [10,20,10,30]
y2 = [20,25,15,25]
plt.bar(x,y1,color='red')
plt.bar(x,y2,bottom=y1,color='blue')
plt.show()
```

```python
x = ['A','B','C','D']
y1 = np.array([10,20,10,30])
y2 = np.array([20,25,15,25])
y3 = np.array([12,15,19,6])
y4 = np.array([10,29,13,19])

plt.bar(x,y1,color='red')
plt.bar(x,y2, bottom=y1,color='black')
plt.bar(x,y3,bottom=y1+y2,color='cyan')
plt.bar(x,y4,bottom=y1+y2+y3,color='green')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1","Round 2","Round 3","Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```
